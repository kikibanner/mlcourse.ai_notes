## Applying Functions to Cells, Columns, and Rows
To apply function to each column, pakenya `apply()`

```python
df.apply(np.max)
```

Method `apply` juga bisa dipake pada tiap row. Untuk melakukannya, spesifikkan pake `axis=1`. Pada skenario ini, fungsi lambda sangat dapat diandalin. COntoh di bawah: select semua states yang diawali dengan 'W':

```python
df[df['State'].apply(lambda state: state[0] == 'W')].head()
```

Method `map` juga bisa dipake buat ngereplace value di kolom dengan ngepass sebuah dictionary dengan format `{old_value: new_value}` sebagai argument-nya

```python
d = {"No": False, "Yes": True}
df["International plan"] = df["International plan"].map(d)
df.head()
```

Hal yang sama bisa dilakuin pake method `replace`. Bedanya kalo replace, kalo ga ketemu yang mau direplace, maka ga diapa2in. Kalo pake `map`, kalo ga ketemu bakal diisi NaN

```python
a_series = pd.Series(['a', 'b', 'c'])
a_series.replace({'a': 1, 'b': 1}) # 1, 2, c
a_series.map({'a': 1, 'b': 2}) # 1, 2, NaN
```

## Grouping
Secara generalm grouping di pandas kayak gini:
> `df.groupby(by=grouping_columns)[columns_to_show].function()`
1. Method `groupby` membagi `grouping_columns` by valuenya. Value yang digrup jadi index.
2. lalu, column of interest diselect (`columns_to_show`). Kalau kita ga nentuin, semua kecuali yang digroupby bakal dishow.
3. Yang terahir, function-nya diapply ke group per selected columns.

```python
# contoh: ngegrup data berdasarkan nilai variabel Churn, dan ngedisplay statistic dari tiga kolom pada tiap grupnya

columns_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']

# df.groupby(by='Churn')['Total day minutes', 'Total eve minutes', 'Total night minutes'].describe(percentiles=[])

df.groupby('Churn')[columns_to_show].describe(percentiles=[])
```

```python
# Melakukan hal yang sama tapi sedikit beda, dengan ngepass sebuah list functions ke agg()

columns_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']

df.groupby('Churn')[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

## Summary Tables
Contoh: kita mau observasi dari distribusi dua buah variabel - `Churn` dan `International plan`. untuk melakukannya, kita bisa buat ***Contingency table*** dengan menggunakan method `crosstab`:

```python
pd.crosstab(df['Churn'], df['International plan'])
```

Dapat dilihat kalau kebanyakan usernya adalah user yang loyal dan tidak menggunakan service tambahan.

Hal tersebut mengingatkan kita dnegan ***pivot table*** pada ms excel. Tentu saja, pandas juga punya method untuk pivot table, yaitu `pivot_table` yang membutuhkan tiga buah parameter:
- `values` - sebuah list dari variable untuk dihitung statistiknya
- `index` - sebuah list dari variables untuk digroupby
- `aggfunc` - statistik apa yang mau dihitung. Misal: sum, mean, max, min, dll

```python
df.pivot_table(
    ['Total day calls', 'Total eve calls', 'Total night calls'],
    ['Area code'],
    aggfunc="mean"
)
```

## Dataframe Transformations

Seperti semua hal yang bisa kita lakuin pake pandas, nambahin kolom ke DataFrame bisa dilakukan dalam berbagai cara.

Sebagai contoh, jika kita mau ngehitung jumlah call untuk semua user, kita bisa membuat series `total_calls` dan ngepaste ke dalam DataFrame;

```python
total_calls = (
    df['Total day calls']
    + df['Total eve calls']
    + df['Total night calls']
    + df['Total intl calls']
)

df.insert(loc=len(df.columns), column='Total calls', value=total_calls)
df.head()
```

```python
# atau bisa juga langsung gausah pake insert
df['Total charge'] = (
    df['Total day charge']
    + df['Total eve charge']
    + df['Total night charge']
    + df['Total intl charge']
)

df.head()
```

Untuk menghapus kolom atau baris pakai method `drop`, ngepass parameter yaitu `axis` (1 kalau mau ngehapus kolom, 0 kalau mau ngehapus baris). Argumen `inplace` buat nentuin kita mau ngeubah DataFrame originalnya apa engga. Dengan `inplace=False`, maka method `drop` tidak dapat mengubah DataFrame existing dan ngereturn DataFrame baru dengan kolomnya atau barisnya yang udah di drop. Kalau `inplace=True` maka akan ngealter DataFrame.

```python
# ngehapus kolom (kolom yang tadi barusan dibikin)
df.drop(["Total charge", "Total calls"], axis=1, inplace=True)

# ngehapus row
df.drop([1, 2]).head()
```

# 2. First Attemp at Predicting telecom Churn
Kita mau ngelihat bagaimana relasi churn rate dengan variable *International plan*. Kita ngelakuinnya pakai `crosstab` dan juga melakukan visual analysis pake seaborn

```python
pd.crosstab(df['Churn'], df['International plan'], margins=True)
```

```python
# some imports to set up plotting
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

# Graphicnya dalam retina format - more sharp and legible (mudah terbaca)
%config InlineBackend.figure_format = 'retina'
```

```python
sns.countplot(x='International plan', hue='Churn', data=df)
```

Dapat dilihat bahwa kalo pake international plan, churn ratenya lebih tinggi. Mungkin aja karena dengan biaya international call yang kurang kekontrol dan rawan konflik, maka menyebabkan ketidakpuasan pelanggan.

Selanjutnya kita bakal melihat feature penting lain - *Customer service calls*. Kita bakal ngelihat summary table dan visualnya

```python
pd.crosstab(df['Churn'], df['Customer service calls'], margins=True)
```

```python
sns.countplot(x='Customer service calls', hue="Churn", data=df)
```

Meskipun ga begitu keliatan pada summary table, namun pada visualnya (di plotnya), kita dapat lihat rate churn banyak di 4 cs calls dan diatasnya.

Selanjutnya kita bakal ngasih binary feature ke DatafRame kita - `Customer service calls > 3`. Dan sekali lagi melihat relasinya dengan churn.

```python
df['Many_service_calls'] = (df['Customer service calls'] > 3).astype("int")


pd.crosstab(df['Many_service_calls'], df['Churn'], margins=True)
```

```python
sns.countplot(x="Many_service_calls", hue="Churn", data=df)
```

Mari bikin contingency table lagi dengan mempertimbangkan *International plan* dan *Many_service_calls* yang barusan dibuat.

```python
pd.crosstab(df['Many_service_calls'] & df['International plan'], df['Churn'])
```

Dari guessing kita ternyata dapat dilihat kalo user bakal churn kalo banyak cs nelpon dan pake international plan. Dari tabel contongency di atas didapatkan 85.8% (salah nebak 464 + 9 kali)
